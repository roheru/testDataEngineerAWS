<h3>Solución</h3>
<p>Para esta solucíion se implemento un script en python</p>

Se descargar el dataset del siguiente enlace <a href="https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/data">Descargar Dataset</a>
<br>
El script en python que lee el csv, limpia las columnas y calcula la nueva columna AVG se encuentra en el siguiente enlace con toda los archivos relacionados a la prueba de Kaggle <a href="https://github.com/roheru/testDataEngineerAWS/tree/main/Ejercicio%20Kaggle">Ejercicio Kaggle</a>

La estructura es la siguiente:
<ul>
    <li>kaggleDataframeJob.py, Script de python</li>
    <li>data_clean.parquet, Archivo parquet de salida</li>
    <li>house-prices/train.csv, csv de entrada</li>
</ul>


A continuación se muestra el script de python.


```python
import pandas as pd
df = pd.read_csv("house-prices/train.csv", sep=",")

#print(df.columns)

for col in df.select_dtypes(include="number").columns:
    df[col].fillna(df[col].mean(), inplace=True)

#calcula el promedio de dos columnas llamadas MSSUBCLASS Y lotFrontage

df['AVG'] = (df['MSSubClass'] + df['LotFrontage']) / 2

print(df.head())
print(df.shape)

#En esta línea guarda como parquet
df.to_parquet("data_clean.parquet", engine='pyarrow', index=False)

```
