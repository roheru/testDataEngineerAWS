<h2>Terraform</h2>
<p>
    En esta sección se visualiza lo siguientes items para realizar un ejercicio de prueba con Terraform, se llevaran a cabo los siguientes pasos:
    <ul>
        <li>Instalación de Terraform</li>
        <li>Creación de recursos</li>
        <ul>
            <li>IAM ROLE</li>
            <ul>
                <li>Permisos List(Todos)</li>
                <li>Permisos Get(Todos)</li>
                <li>Replicación de Etiquetas</li>
                <li>Permisos de creación y eliminación de bucket</li>
                <li>Sin permisos de lectura a objetos almacenados</li>
            </ul>
            <li>Bucket S3</li>
            <ul>
                <li>ACL: Privada</li>
                <li>Versionamiento habilitado</li>
                <li>Ciclo de Vida: 15 días</li>
            </ul>
            <li>Instancia EC2</li>
        </ul>  
    </ul>
</p>


<p>Para la instalación de Terraform en mac se debera tener en cuenta lo siguientes comandos y acciones</p>
<ul>
    <li>brew tap hashicorp/tap</li>
    <li>brew install hashicorp/tap/terraform</li>
    <li>terraform -v </li>
</ul>

<p>Para la instalación de AWS CLI es necesaria, ya que nos permitirá configurar las variables de entorno de la cuenta de AWS</p>
<ul>
    <li>aws configure</li>
</ul>

Se visualizarán las siguientes configuraciones
<ul>
    <li>AWS Access Key ID [None]: XXXXXXXX_accesss_key_id_de_la_cuenta</li>
    <li>AWS Secret Access Key [None]: XXXXXXXX_access_secret_key_de_la_cuenta</li>
    <li>Default region name [None]: us-east-1</li>
    <li>Default output format [None]: json</li>
</ul>

Para buenas practicas se recomienda separar los archivos de la estructura de Terrraform
<ul>
    <li>terraform-aws-project/</li>
    <ul>
        <li>main.tf</li>
        <li>variables.tf</li>
        <li>outputs.tf</li>
        <li>provider.tf</li>
        <li>README.md</li>
    </ul>
</ul>
